# Social Network Analysis





Using the provided lesmiserables.gml file, do the following:

Visualize the graph. {20 points}


What can you observe from the visualization? What are the THREE most interesting relationships (visually) and why? {20 points}

```python
#importing NetworkX modules
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import networkx as nx

#importing gml dataset
g = nx.read_gml('lesmiserables.gml')

# taking a look at the nodes in gml dataset
g.nodes()
```

```python
#seeing more info on teh dataset
print(g)
```

```python
#looking at edges
g.edges(data=True)
```

```python
#rendering the network

plt.figure(figsize=(16, 16))
nx.draw_networkx(g,
                 node_color='lightgrey',
                 font_color='red',
                 node_size=800)
plt.title('Les Miserables - Social Network')

#limits = plt.axis('off')  #added later
plt.show()
```

```python
# checking out the adjacency matrix
#adj_matrix = nx.adj_matrix(g)      from the book
#print (adj_matrix.todense())

nx.adjacency_matrix(g).todense()
```

```python
#alternative representation for the matrix as an edge list
edge_list=nx.to_pandas_edgelist(g)
print(edge_list)
```

```python
# checking out adjacency list
for n in g.adjacency():
    print(n)
```

# Graph Characteristics


For each of the three persons identified in the previous question, provide the following metrics: {48 points}

- Degree Centrality

- Eigenvector Centrality

- PageRank

- Betweenness




Based on the above metrics, do you believe the three characters you picked are indeed the most central? Is there a character you missed in the visualization that perhaps is a more central figure? If so, whom? {12 points}

```python
#characters identified
characters = ['Valjean', 'Cosette', 'Javert']

#calculating degree centrality
degree_centrality = nx.degree_centrality(g)
nx.set_node_attributes(g, degree_centrality, 'degree_centrality')
deg_values = sorted(set(degree_centrality.values()), reverse=True)
print("Degree Centrality:")
for char in characters:
    print(f"{char}: {degree_centrality[char]}")

#calculating eigenvector centrality
eigenvector_centrality = nx.eigenvector_centrality(g)
nx.set_node_attributes(g, eigenvector_centrality, 'eigenvector_centrality')
eig_values = sorted(set(eigenvector_centrality.values()), reverse=True)
print("\nEigenvector Centrality:")
for char in characters:
    print(f"{char}: {eigenvector_centrality[char]}")

#calculating PageRank
pagerank = nx.pagerank(g)
nx.set_node_attributes(g, pagerank, 'pagerank')
pr_values = sorted(set(pagerank.values()), reverse=True)
print("\nPageRank:")
for char in characters:
    print(f"{char}: {pagerank[char]}")

#calculating Betweenness
betweenness = nx.betweenness_centrality(g)
nx.set_node_attributes(g, betweenness, 'betweenness')
bet_values = sorted(set(betweenness.values()), reverse=True)
print("\nBetweenness:")
for char in characters:
    print(f"{char}: {betweenness[char]}")
```
